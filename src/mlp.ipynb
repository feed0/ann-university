## Etapa 01 - Multi-Layer Perceptron

    - MLP: Rede Multi-Layer Perceptron

### Instruções:

1. Acesse a documentação do Multi-Layer Perceptron via link: 
    * https://scikit-learn.org/stable/modules/generated/sklearn.neural_network.MLPClassifier.html
2. Estude os seguintes parâmetros:
    * 'activation' : Função de ativação para a camada oculta ({'identity', 'logistic', 'tanh', 'relu'}).
    * 'solver' : O solucionador para otimização de peso {'lbfgs', 'sgd', 'adam'}. São métodos para a minimização de erros pela descida do gradiente.
    * 'learning_rate_init' : A taxa de aprendizagem controla o tamanho do passo na atualização dos pesos. Usado apenas quando solver='sgd' ou 'adam'.
    * 'max_iter' : Número máximo de iterações. O solucionador itera até a convergência (determinada por 'tol') ou este número de iterações (epochs).
    * 'tol' : Tolerância para a otimização.
    * 'momentum' : Momento para atualização de descida gradiente. Deve estar entre 0 e 1. Usado apenas quando solucionador='sgd'.
    * 'Shuffle' : Se as amostras devem ser embaralhadas em cada iteração. Usado apenas quando solucionador='sgd' ou 'adam'.
    * 'random_state' : Tornar reproduzível a geração de números aleatórios para inicialização de pesos e polarizações para treino e teste.

    Bibliotecas

```python
%pip install -U scikit-learn
from sklearn import datasets
import pandas as pd
from sklearn.neural_network import MLPClassifier
from sklearn.model_selection import GridSearchCV, train_test_split
from sklearn.metrics import precision_score, recall_score, f1_score
from sklearn.metrics import classification_report
import matplotlib.pyplot as plt
```

    Definindo a base de dados (dataset) IRIS

```python
from sklearn import datasets

iris = pd.DataFrame(datasets.load_iris().data, columns=datasets.load_iris().feature_names)
iris
```

    Atribuindo o dataset para as entradas e saídas

```python
from sklearn.datasets import load_iris

iris = load_iris()
inputs  = iris.data
print("Entradas: ", inputs.shape)
outputs = iris.target
print("Saídas: ", outputs.shape)
```

    Criando a RNA Multi-Layer Perceptron

Divisão do Conjunto de Dados em Treinamento e Teste.


```python
inputs_train, inputs_test, outputs_train, outputs_test = train_test_split(inputs, outputs, test_size=0.2, random_state=42)
```

```python
print(f"Tamanho do conjunto de treino: {inputs_train.shape}, Tamanho do conjunto de teste: {inputs_test.shape}")
```

```python
from sklearn.neural_network import MLPClassifier # Multi-Layer Perceptron
```

    Treinamento da MLP

```python
mlp = MLPClassifier(hidden_layer_sizes=(10,), max_iter=1000, random_state=42)
mlp.fit(inputs_train, outputs_train) 
```

    Realizar previsões pela MLP treinada

```python
predictions = mlp.predict(inputs_test)
```

    Mapeamento de classes

```python
def map_predictions(predictions):
    classes = {0: "Setosa", 1: "Versicolor", 2: "Virginica"}
    return [classes[pred] for pred in predictions]

predictions = mlp.predict(inputs_test)
predicted_classes = map_predictions(predictions)
print(predicted_classes)
```

    Divisão da base de dados: 80/20 e 70/30

```python
from sklearn.model_selection import train_test_split, GridSearchCV

# Dividir o conjunto de dados em 80% treino e 20% teste
X_train_80, X_test_20, y_train_80, y_test_20 = train_test_split(inputs, outputs, test_size=0.2, random_state=42)

# Dividir o conjunto de dados em 70% treino e 30% teste
X_train_70, X_test_30, y_train_70, y_test_30 = train_test_split(inputs, outputs, test_size=0.3, random_state=42)

```

```python
from sklearn.metrics import precision_score, recall_score, f1_score, classification_report

# Para o conjunto de treino 80/20
mlp.fit(X_train_80, y_train_80)
predictions_80 = mlp.predict(X_test_20)
print("Métricas para o conjunto de treino 80/20:")
print(classification_report(y_test_20, predictions_80))

# Para o conjunto de treino 70/30
mlp.fit(X_train_70, y_train_70)
predictions_70 = mlp.predict(X_test_30)
print("Métricas para o conjunto de treino 70/30:")
print(classification_report(y_test_30, predictions_70))
```
